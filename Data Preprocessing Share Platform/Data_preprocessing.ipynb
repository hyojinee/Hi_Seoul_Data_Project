```
import tkinter as tk
from tkinter import ttk
from tkinter.filedialog import askopenfilename
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter   
from tkinter import messagebox
import numpy as np

%matplotlib inline
%matplotlib qt

# --- classes ---

class MyWindow:
    
    #시작 창
    def __init__(self, parent):
            
            self.parent = parent
            
            #  데이터 정보 및 불러오기 frame1
            self.frame1 = tk.Frame(self.parent,bg="white")
            

            self.filename = None
            self.df = None

            self.text = tk.Text(self.frame1,highlightthickness=4, relief = 'raised')
            self.text.pack(padx = 2, pady = 1)
            
            self.button = tk.Button(self.frame1, text='데이터 불러오기',bg='#0D1D3C', fg="white", command=self.load)
            self.button.pack()
            
            self.button = tk.Button(self.frame1, text='데이터 정보보기',bg='#0D1D3C', fg="white", command=self.display)
            self.button.pack()
            
            self.button = tk.Button(self.frame1, text='boxplot 보기', bg='#0D1D3C', fg="white", command=self.boxplot)
            self.button.pack()
            
            self.button = tk.Button(self.frame1, text='데이터 저장하기', bg='#0D1D3C', fg="white",command=self.update)
            self.button.pack()
            
            self.frame1.pack()
            
            # 결측치 처리 frame2 
            self.frame2 = tk.Frame(self.frame1,bg="white")

            self.label = tk.Label( self.frame2 , bg="white",text = '결측치 처리', font=('family',25,'bold'))
            self.label.pack()
            
            # 결측치 체크버튼
            self.var_1 = tk.IntVar()
            self.var_2 = tk.IntVar()
            self.var_3 = tk.IntVar()
            self.var_4 = tk.IntVar()
            self.var_5 = tk.IntVar()
            self.var_6 = tk.IntVar()

            self.book_1 = tk.Checkbutton( self.frame2 ,variable = self.var_1 ,bg="white",font=('family',20,'bold'),text = '평균값' , command = self.fill_mean )
            self.book_2 = tk.Checkbutton( self.frame2 ,variable = self.var_2 ,bg="white",font=('family',20,'bold'),text = '중간값' , command = self.fill_median  )
            self.book_3 = tk.Checkbutton( self.frame2 ,variable = self.var_3 ,bg="white",font=('family',20,'bold'),text = '최소값' , command = self.fill_min  )
            self.book_4 = tk.Checkbutton( self.frame2 ,variable = self.var_4 ,bg="white",font=('family',20,'bold'),text = '최대값' , command = self.fill_max  )
            self.book_5 = tk.Checkbutton( self.frame2 ,variable = self.var_5 ,bg="white",font=('family',20,'bold'),text = '0값   ' , command = self.fill_zero )
            self.book_6 = tk.Checkbutton( self.frame2 ,variable = self.var_6 ,bg="white",font=('family',20,'bold'),text = '삭제(비추천)' , command = self.drop_null ) 

            self.book_1.pack( side = tk.BOTTOM)
            self.book_2.pack( side = tk.BOTTOM)
            self.book_3.pack( side = tk.BOTTOM)
            self.book_4.pack( side = tk.BOTTOM)
            self.book_5.pack( side = tk.BOTTOM)
            self.book_6.pack( side = tk.BOTTOM)
            

            self.frame2.pack( side="left", fill="both", padx = 10, pady = 30 )
            
            # 이상치 처리 frame3 
            self.frame3 = tk.Frame(self.frame1,bg="white")
            

            self.label = tk.Label(self.frame3 , text = '이상치 처리',bg="white", font=('family',25,'bold'))
            self.label.pack()
            
            # 이상치 체크버튼
            self.var_7 = tk.IntVar()
            self.book_7 = tk.Checkbutton( self.frame3 , variable = self.var_7,bg="white", font=('family',20,'bold'),text = '제거' , command = self.dialog )
            self.book_7.pack( side = tk.TOP)

            self.frame3.pack( side="right", fill="both", padx = 10, pady = 30 )

    # 데이터 불러오기  
    def load(self):
            self.name = askopenfilename(filetypes=(("Excel files", "*.xlsx"),("All files", "*.*") ))
            if self.name:
                if self.name.endswith('.csv'):
                    self.df = pd.read_csv(self.name)
                else:
                    self.df = pd.read_excel(self.name)
                    
                self.filename = self.name
                
    # 데이터 정보보기     
    def display(self):
        if self.df is None:
            self.load()
           # display if loaded
        if self.df is not None:
            self.text.insert('end', self.filename + '\n')
            self.text.insert('end', '---------------------- 데이터 정보 ----------------------\n')
            self.text.insert('end', 'Describe: \n' + str(self.df.describe()) + '\n\n')
            self.text.insert('end', '데이터 summary: \n'+ str(self.df.head()) + '\n\n')
            self.text.insert('end', '컬럼별 결측치 개수: \n'+ str(self.df.isnull().sum()) + '\n\n')
            sns.heatmap(self.df.isnull(), cbar = False)
            self.df.plot()
           
            
    # 데이터 이상치 그래프 보기        
    def boxplot(self):
        if self.df is None:
            self.load()
        if self.df is not None:
            self.text.insert('end', self.filename + '\n')
            self.text.insert('end', '---------------------- box plot ----------------------\n')
            self.df.boxplot()   
            
    # 데이터 저장하기           
    def update(self) :
        self.update_data = self.df
        self.update_data.to_excel(excel_writer='update.xlsx')
        self.text.insert('end', '**********************데이터가 저장되었습니다********************\n')
        

    #-----------------결측치 처리 함수-------------------
    
    def fill_mean(self):
            # 결측치 평균값으로 채우기
            self.df.set_index('Name',inplace = True)
            self.df.reset_index(inplace = True)
            self.fill_mean = self.df.fillna(self.df.mean(),axis=0)
            
            self.text.insert('end', '---------------------- 결측치 처리 후 ----------------------\n')
            self.text.insert('end', '데이터 summary: \n'+ str(self.fill_mean.head()) + '\n\n')
            self.text.insert('end', '컬럼별 결측치 개수: \n'+ str(self.fill_mean.isnull().sum()) + '\n\n')
            sns.heatmap(self.fill_mean.isnull(), cbar = False)
            
            self.df = self.fill_mean
            return(self.df)

    def fill_median(self):
            # 결측치 중간값으로 채우기
            self.df.set_index('Name',inplace = True)
            self.df.reset_index(inplace = True)
            self.fill_median = self.df.fillna(self.df.median(),axis=0)
            
            self.text.insert('end', '---------------------- 결측치 처리 후 ----------------------\n')
            self.text.insert('end', '데이터 summary: \n'+ str(self.fill_median.head()) + '\n\n')
            self.text.insert('end', '컬럼별 결측치 개수: \n'+ str(self.fill_median.isnull().sum()) + '\n\n')
            sns.heatmap(self.fill_median.isnull(), cbar = False)
            
            self.df = self.fill_median
            return(self.df)
            

    def fill_min(self):
            # 결측치 최소값으로 채우기
            self.df.set_index('Name',inplace = True)
            self.df.reset_index(inplace = True)
            self.fill_min = self.df.fillna(self.df.min(),axis=0)
            
            self.text.insert('end', '---------------------- 결측치 처리 후 ----------------------\n')
            self.text.insert('end', '데이터 summary: \n'+ str(self.fill_min.head()) + '\n\n')
            self.text.insert('end', '컬럼별 결측치 개수: \n'+ str(self.fill_min.isnull().sum()) + '\n\n')
            sns.heatmap(self.fill_min.isnull(), cbar = False)
            
            self.df = self.fill_min
            return(self.df)

    def fill_max(self):
            # 결측치 최대값으로 채우기
            self.df.set_index('Name',inplace = True)
            self.df.reset_index(inplace = True)
            self.fill_max = self.df.fillna(self.df.max(),axis=0)
            
            self.text.insert('end', '---------------------- 결측치 처리 후 ----------------------\n')
            self.text.insert('end', '데이터 summary: \n'+ str(self.fill_max.head()) + '\n\n')
            self.text.insert('end', '컬럼별 결측치 개수: \n'+ str(self.fill_max.isnull().sum()) + '\n\n')
            sns.heatmap(self.fill_max.isnull(), cbar = False)
            
            self.df = self.fill_max
            return(self.df)

    def fill_zero(self):
            # 결측치 0값으로 채우기
            self.df.set_index('Name',inplace = True)
            self.df.reset_index(inplace = True)
            self.fill_zero = self.df.fillna(0)
            
            self.text.insert('end', '---------------------- 결측치 처리 후 ----------------------\n')
            self.text.insert('end', '데이터 summary: \n'+ str(self.fill_zero.head()) + '\n\n')
            self.text.insert('end', '컬럼별 결측치 개수: \n'+ str(self.fill_zero.isnull().sum()) + '\n\n')
            sns.heatmap(self.fill_zero.isnull(), cbar = False)
            
            self.df = self.fill_zero
            return(self.df)

            
    def drop_null(self):
             # 결측치 있는 컬럼 삭제
            self.df.set_index('Name',inplace = True)
            self.df.reset_index(inplace = True)
            self.drop_null = self.df.dropna(axis = 0, how='any').reset_index(drop= True)
            
            self.text.insert('end', '---------------------- 결측치 처리하기 후 ----------------------\n')
            self.text.insert('end', '데이터 summary: \n'+ str(self.drop_null.head()) + '\n\n')
            self.text.insert('end', '컬럼별 결측치 개수: \n'+ str(self.drop_null.isnull().sum()) + '\n\n')
            sns.heatmap(self.drop_null.isnull(), cbar = False)
            
            self.df = self.drop_null
            return(self.df)
               

        
   # 이상치 처리 삭제 체크박스 체크 시 새로운 창 만들기
    def dialog(self):
        
        self.text.insert('end', '---------------------- 이상치 처리 후 ----------------------\n')

        self.column_index = tk.Tk()
        self.column_index.title("Outliers processing page")
        self.canvas1 = tk.Canvas(self.column_index, width = 400, height = 300,  relief = 'raised',bg = "white")

        self.label2 = tk.Label(self.column_index, text='컬럼명을 입력하세요:')
        self.label2.config(font=('family', 30))
        self.label2.pack()
        self.canvas1.create_window(200, 100, window=self.label2)

        self.entry1 = tk.Entry(self.canvas1)
        self.canvas1.create_window(200, 140, window= self.entry1)
        
        self.button1 = tk.Button(self.column_index, text='삭제하기', bg='#0D1D3C',fg="white", command =self.removeOutliers)
        self.canvas1.create_window(200, 200, window=self.button1)
        
        self.outlier = tk.Text(self.column_index)
        self.outlier.pack(padx = 2, pady = 1)
        
        self.canvas1.pack()
    
    # 이상치 삭제할 컬럼명 입력 
    def removeOutliers(self):
        
        self.column = self.entry1.get()
    
        self.label11 = tk.Label(self.column_index, text= '삭제할 컬럼명:' + self.column, font=('family', 20))
        self.canvas1.create_window(200, 220, window=self.label11)
        self.canvas1.pack()
        
        self.column = int(self.column)
        
        #이상치 제거 실행 
        self.q1 = self.df[self.column].quantile(0.25)
        self.q3 = self.df[self.column].quantile(0.75)

        # 1.5 * IQR(Q3 - Q1)
        self.iqt = 1.5 * (self.q3 - self.q1)

        # 원래 데이터 복제
        self.y = self.df.copy()

        #이상치를 NA로 변환
        self.y[self.column][(self.df[self.column] > (self.q3 + self.iqt)) | (self.df[self.column] < (self.q1 - self.iqt))] = None

        self.y = self.y.dropna(axis=0)

        #print("이상치 제거" ,self.df)
#         self.outlier.insert('end', str(max(self.y) + '\n')
        self.outlier.insert('end', str(self.y.head())  + '\n')                  
        self.outlier.insert('end', '---------------------- box plot ----------------------\n')
        self.y.boxplot()
 
                            



# --- main ---

if __name__ == '__main__':
    root = tk.Tk()
    top = MyWindow(root)
    root.title("Data preprocessing page")
    root.mainloop()
```
